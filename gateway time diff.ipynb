```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gw1 = open("/Users/gexin/Desktop/gw1.txt","r")
gw2 = open("/Users/gexin/Desktop/gw2.txt","r")
```

```python
gw1_data = []
gw2_data = []
for line in gw1:
    gw1_data.append(line.split(" "))
for i in range(len(gw1_data)):
    gw1_data[i][4] = int(gw1_data[i][4][:-1])
    gw1_data[i][0] = int(gw1_data[i][0])
    gw1_data[i][1] = int(gw1_data[i][1])
    gw1_data[i][3] = int(gw1_data[i][3]) 
    #print(gw1_data[i])
    
for line in gw2:
    gw2_data.append(line.split(" "))
for i in range(len(gw1_data)):
    gw2_data[i][4] = int(gw2_data[i][4][:-1])
    gw2_data[i][0] = int(gw2_data[i][0])
    gw2_data[i][1] = int(gw2_data[i][1])
    gw2_data[i][3] = int(gw2_data[i][3]) 
    #print(gw1_data[i])
```

```python
err = []
tot = 0.0
for i in range(len(gw1_data)):
    for j in range(len(gw2_data)):
        if gw1_data[i][3] == gw2_data[j][3] and gw1_data[i][1] == gw2_data[i][1]:
            if np.abs(gw1_data[i][4] - gw2_data[j][4])/1000.0 < 10.0 :
                err.append((gw1_data[i][4] - gw2_data[j][4])/1000.0)
                tot += np.abs((gw1_data[i][4] - gw2_data[j][4])/1000.0)
print(len(err))
print("E(error) = " + str(tot/len(err)))
```

```python
plt.figure(figsize=(8, 6), dpi=100) 
plt.xlabel('number of pkts')  
plt.ylabel('error on test 2 (us)')  
x = [i for i in range(len(err))]  
plt.plot(x,err,"+",label="error")  
plt.grid(True)    
plt.show()  
```
