```python
import numpy as np
import time
```

```python
time_v = []
err = []
maxerr = []
z = []
for i in range(2,1001):
    a = np.loadtxt("/Users/gexin/Desktop/data_1/"+str(i)+".txt",delimiter=" ")
    
    start = time.clock()
    b = np.linalg.inv(a)
    end = time.clock()
    
    cur_err=0.0
    cur_maxerr=0.0
    c = np.dot(a,b)
    for j in range(0,i):
        for k in range(0,i):
            if j!=k:
                cur_err += np.abs(c[j][k])
                cur_maxerr = max(cur_maxerr, np.abs(c[j][k]))
    err.append(cur_err)
    maxerr.append(cur_maxerr)
    time_v.append(end-start)
    print("#"+str(i)+": time: "+str(end-start)+" err: "+str(cur_err)+" maxerr: "+str(cur_maxerr))
    
z = [str(data) for data in time_v]
np.savetxt("/Users/gexin/Desktop/result/py_1_time.txt", z, fmt="%s", delimiter=" ")
np.savetxt("/Users/gexin/Desktop/result/py_1_err.txt", err, fmt="%s", delimiter=" ")
np.savetxt("/Users/gexin/Desktop/result/py_1_maxerr.txt", maxerr, fmt="%s", delimiter=" ")
print("done")
```
