```python
tiny = 1e-6
for i in range(2,1001):
    a = np.loadtxt("/Users/gexin/Desktop/data_1/"+str(i)+".txt",delimiter=" ")
    x = np.random.randint(1,i)
    a[0] = a[x]
    a[0][x] += tiny
    np.savetxt("/Users/gexin/Desktop/data_2/"+str(i)+".txt", a, fmt="%s", delimiter=" ")
    print("size: "+str(i)+" done")
```
