```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
node1 = open("/Users/gexin/Downloads/lora1.txt","r");
gw1_data = []
for line in node1:
    gw1_data.append(line.split())
    
plt.figure(figsize=(8, 6), dpi=120) 
plt.xlabel('received time(s) of pkt')  
plt.ylabel('rssi')  
x = []
y = []
tot = 0
for i in range(len(gw1_data)):
    if int(gw1_data[i][2]) < 3000: 
        x.append(gw1_data[i][2])
        y.append(gw1_data[i][4])
plt.plot(x,y,"+")  
plt.grid(True)    
plt.show()

num = 0;
for i in y:
    if int(i) >= -40:
        num+=1

t = int(x[-1])/10

print("pkt loss: "+str(1.0-num*1.0/t))
```

```python
node3 = open("/Users/gexin/Downloads/loramix.txt","r");
gw3_data = []
for line in node3:
    gw3_data.append(line.split())
    
plt.figure(figsize=(8, 6), dpi=120) 
plt.xlabel('received time(s) of pkt')  
plt.ylabel('rssi')  
x1 = []
x2 = []
y1 = []
y2 = []
num = 0
for i in range(len(gw3_data)):
    if gw3_data[i][1] == "01E67CE6" and int(gw3_data[i][2]) < 3000:
        x1.append(gw3_data[i][2])
        y1.append(gw3_data[i][4])  
    elif int(gw3_data[i][2]) < 3000:
        x2.append(gw3_data[i][2])
        y2.append(gw3_data[i][4])
    
plt.plot(x1,y1,"+")  
plt.plot(x2,y2,"*")
plt.grid(True)    
plt.show()

num1 = 0
num2 = 0
for i in y1:
    if int(i) >= -40:
        num1+=1
for i in y2:
    if int(i) >= -40:
        num2+=1        
t = max(int(x1[-1]),int(x2[-1]))/10.0
print("tot pkt loss: " +str(1-(num1+num2)/t/2.0))
print("node1 pkt loss: " + str(1.0-num1/t))
print("node2 pkt loss: " + str(1.0-num2/t))
```
