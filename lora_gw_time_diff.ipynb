```python
gw1 = open("/Users/gexin/Desktop/loraall/test_data/gps_short1.txt","r")
gw2 = open("/Users/gexin/Desktop/loraall/test_data/gps_short2.txt","r")

gw1_time = []
gw2_time = []

for line in gw1:
    if len(line) > 10 and line[0]=='>' and line[1]=='>' and line[2]=='n' and line[3]=='s' and line[4]==':':
        gw1_time.append(line[5:len(line)-1])
        
for line in gw2:
    if len(line) > 10 and line[0]=='>' and line[1]=='>' and line[2]=='n' and line[3]=='s' and line[4]==':':
        gw2_time.append(line[5:len(line)-1])
        
print("gw1 len: " +str(len(gw1_time)))
print("gw2 len: " +str(len(gw2_time)))

data1 = [int(i) for i in gw1_time]
x1 = [i for i in range(len(data1))]
data2 = [int(i) for i in gw2_time]
x2 = [i for i in range(len(data2))]   
plt.figure(figsize=(10, 8), dpi=200) 
plt.xlabel('short packet')  
plt.ylabel('RX(GPS) time ')  
plt.plot(x1,data1,"+",label="python error") 
plt.plot(x2,data2,"*",label="python error")  
plt.grid(True)  
plt.show()
```
