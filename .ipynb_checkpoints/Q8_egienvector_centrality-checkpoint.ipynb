```python
import numpy as np
import matplotlib.pyplot as plt
```

### 将图的信息存入links列表

```python
links = [[1,3],[3,6],[3,8],[3,7],[1,2],[1,4],[2,4],[2,5],[4,5],[6,7],[7,8]]
```

### 建立元素全部为0的 8*8矩阵

```python
map = np.zeros([8,8]) #size = 9 * 9
```

### 将link信息存入图里

```python
for link in links:
    map[link[0]-1][link[1]-1] = 1
    map[link[1]-1][link[0]-1] = 1
print(map)
```

### 计算特征值和特征向量，存入eVal, eVec

```python
eVal, eVec = np.linalg.eig(map)
```

$$x(t) = A^tx(0)$$ 
$$x(0) = \sum_{i}c_iV_i$$
可以得到 $$x(t) = A^t\sum_{i}c_iV_i = \sum_{i}c_iA^tV_i = \sum_{i}c_i\lambda ^t V_i$$

<script type="text/javascript" src="http://cdn.mathjax.org/mathjax/latest/MathJax.js?config=default"></script>
其中 $\{c_i\}$为权重，$\{\lambda_i\}$为特征值，$\{V_i\}$为特征向量

当 $t\to \infty$时，$\lambda_1$ "**will dominate**"
所以可以只观察$\lambda_1$的影响，$Ax = \lambda_1x$


意味着 $$x_i = \frac{1}{\lambda_1} \sum_{j}A_{ij}x_j, \forall i$$

将$X$正则化(normonlize)后，即得到我们需要的 egienvector centrality

```python
degree      = np.array([0.375,0.375,0.500,0.375,0.250,0.250,0.375,0.250])
closeness   = np.array([0.636,0.500,0.636,0.500,0.368,0.438,0.467,0.438])
betweenness = np.array([0.685,0.143,0.713,0.029,0.000,0.000,0.029,0.000])
eigenvector = np.array([0.406,0.346,0.476,0.346,0.241,0.295,0.371,0.295])

x = [5,6,3,6,7,2,1,2]
y = [0,1,0,-1,0,1,0,-1]

plt.axes(facecolor=(1.0,1.0,1.0))
plt.scatter(x, y, s=eigenvector**3*40000)
plt.show()
```

#### 初始化单位向量
$v = [1,1,...,1]$

设向量p为A的最大特征值对应的特征向量

那么误差为 $err = || p-v ||_2$

随着迭代 $v = A\frac{v}{||v||_2}$

$err$不断的变小，向量$v$被矩阵$A$变换成了向量$p$


```python
v = np.ones(8)
err = []
maxerr = 1e9;
step = 0
while maxerr>=1e-8:
    step += 1
    p = np.sum(np.abs(v-eVec.T[0]))
    maxerr = p
    err.append(p)
    v = np.dot(map,v)
    v = v/np.sqrt(np.dot(v,v.T))
err.remove(err[0])

plt.figure(figsize=(8, 6), dpi=100) 
plt.xlabel('steps')  
plt.ylabel('error')  
x = [i for i in range(1,step)] 
plt.plot(x,err,"*",label="python error")  
plt.grid(True)  
plt.show()  
```
